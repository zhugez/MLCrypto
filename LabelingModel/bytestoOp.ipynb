```python
import os
from evmdasm import EvmBytecode
import json

def process_bytecode_file(file_path, output_dir):
    # Read the bytecode from file
    with open(file_path, 'r') as f:
        evm_bytecode = f.read().strip()
    
    try:
        # Disassemble the bytecode
        evmcode = EvmBytecode(evm_bytecode)
        evminstructions = evmcode.disassemble()
        
        # Extract opcodes
        opcodes = [instr.name for instr in evminstructions]
        
        # Create output filename
        base_name = os.path.splitext(os.path.basename(file_path))[0]
        output_file = os.path.join(output_dir, f"{base_name}_opcodes.json")
        
        # Save opcodes to JSON file
        with open(output_file, 'w') as f:
            json.dump({
                "source_file": file_path,
                "opcodes": opcodes
            }, f, indent=2)
            
        return True
    except Exception as e:
        print(f"Error processing {file_path}: {str(e)}")
        return False

def main():
    # Define directories
    dataset_dirs = [
        "./output/reentrancy/",
        "./output/gaslimit/",
        "./output/integeroverflow/",
    ]
    
    output_dirs = [
        "./OutputOpcodes/reentrancy/",
        "./OutputOpcodes/gaslimit/",
        "./OutputOpcodes/integeroverflow/",
    ]
    total_files = 0
    successful_files = 0
      
    # Walk through the Dataset directory
    for i in range(len(dataset_dirs)):
        # Create output directory if it doesn't exist
        os.makedirs(output_dirs[i], exist_ok=True)
        
        for root, _, files in os.walk(dataset_dirs[i]):
            for file in files:
                if file.endswith('.txt') or file.endswith('.bin'):
                    total_files += 1
                    file_path = os.path.join(root, file)
                    
                    if process_bytecode_file(file_path, output_dirs[i]):  # Pass the corresponding output directory
                        successful_files += 1

    print(f"Processed {successful_files} out of {total_files} files successfully")
if __name__ == "__main__":
    main()
```
