```python
from evmdasm import EvmBytecode

# Convert bytecode to opcode
bytecodes = df['bytecode']
evm_bytecodes = [EvmBytecode(bytecode) for bytecode in bytecodes]
opcodes = [evm_bytecode.disassemble() for evm_bytecode in evm_bytecodes]
opcodes[:5]
# data = pd.DataFrame(opcodes, columns=['opcodes'])
# data.head()
```

```python
instruction_names = [[instr.name for instr in opcode] for opcode in opcodes]
data = pd.DataFrame({'opcodes': instruction_names})
data.head()
```
